```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
response = urlopen('https://en.wikipedia.org/wiki/Main_Page')
soup = BeautifulSoup(response, 'html.parser') #response를 soup에 넣는다
for anchor in soup.find_all('a'): # anchor라는 변수에 넣는다
    print(anchor.get('href','/'))
```

```python
response = urlopen('https://www.naver.com/')
soup = BeautifulSoup(response, 'html.parser') #response를 soup에 넣는다
for anchor in soup.select('span.ah_k'):
    print(anchor)
```

```python
response = urlopen('https://www.naver.com/')
soup = BeautifulSoup(response, 'html.parser') #response를 soup에 넣는다
for anchor in soup.select('span.ah_k'):
    print(anchor.get_text())
```

```python
response = urlopen('https://www.naver.com/')
soup = BeautifulSoup(response, 'html.parser') #response를 soup에 넣는다
i = 1
for anchor in soup.select('span.ah_k'):
    print(str(i) + '위: '+ anchor.get_text())
    i += 1
```

```python
# Google: PYthon write text file

response = urlopen('https://www.naver.com/')
soup = BeautifulSoup(response, 'html.parser') #response를 soup에 넣는다
i = 1
f = open('C:/codeit/새파일.txt', 'w')

for anchor in soup.select('span.ah_k'):
    data = str(i) + '위: '+ anchor.get_text() + '\n'
    i += 1
    f.write(data)
f.close()

# 파일 생성
```
