# 데이터 합치기

- inner join : 겹치는 부분만 합침
- left outer join: 왼쪽 데이터 프레임에 있는 부분만 합침
- right outer join: 오른쪽 부분에 존재하는 부분만 합침
- full outer join: 양쪽 모든 부분을 합침


```python
price_df = pd.read_csv('C:/codeit/vegetable_price.csv')
quantity_df = pd.read_csv('C:/codeit/vegetable_quantity.csv')
```

```python
price_df
```

```python
quantity_df
```

```python
pd.merge(price_df, quantity_df, on='Product')
# 겹치는 부분만 나옴
```

```python
pd.merge(price_df, quantity_df, on='Product', how = 'left')
```

```python
pd.merge(price_df, quantity_df, on='Product', how = 'right')
```

```python
pd.merge(price_df, quantity_df, on='Product', how = 'outer')
```
