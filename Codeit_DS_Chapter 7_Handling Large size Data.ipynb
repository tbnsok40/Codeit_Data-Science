# Q. 강의실 배정하기 pt.1

```python
df = pd.read_csv('C:/codeit/enrolment_2.csv')

df['room assignment'] = 'not assigned'


allowed = df['status'] == 'allowed'
course_count = df.loc[allowed, 'course name'].value_counts()
# df 문에 조건 넣을 땐, 무조건 .loc 달아줘야 한다.


aud = list(course_count[course_count >= 80].index)
lar  = list(course_count[(80 > course_count)&(course_count >= 40)].index)
medi = list(course_count[(40 > course_count)&(course_count >= 15)].index)
small = list(course_count[(15 > course_count)&(course_count >= 5)].index)



for course in aud:
    df.loc[(df['course name']==course) & allowed, 'room assignment'] = 'Auditorium'

for course in lar:
    df.loc[(df['course name']==course) & allowed, 'room assignment'] = 'Large room'

for course in medi:
    df.loc[(df['course name']==course) & allowed, 'room assignment'] = 'Medium room'

for course in small:
    df.loc[(df['course name']==course) & allowed, 'room assignment'] = 'Small room'

df
    
# 포문을 돌려서, 나오는 인덱스가 aud에 속하면, room assignment 에 auditorium을 할당
```

# Q. 강의실 배정하기 pt.2

```python
df = pd.read_csv('C:/codeit/enrolment_3.csv')
df.rename(columns = {'room assignment':'room number'}, inplace = True)

allowed = df['status'] == 'allowed'
course_count = df.loc[allowed, 'course name'].value_counts()
# df 문에 조건 넣을 땐, 무조건 .loc 달아줘야 한다.

aud = list(course_count[course_count >= 80].index)
lar  = list(course_count[(80 > course_count)&(course_count >= 40)].index)
medi = list(course_count[(40 > course_count)&(course_count >= 15)].index)
small = list(course_count[(15 > course_count)&(course_count >= 5)].index)

aud.sort()
lar.sort()
medi.sort()
small.sort()

df.rename(columns = {'room assignment':'room number'})

number = 1
for course in aud:
    df.loc[df['course name']==course, 'room number'] = 'Auditorium-' + str(number)
    number += 1

number = 1
for course in lar:
    df.loc[df['course name']==course, 'room number'] = 'Large-' + str(number)
    number += 1

number = 1
for course in medi:
    df.loc[df['course name']==course, 'room number'] = 'Medium-' + str(number)
    number += 1

number = 1
for course in small:
    df.loc[df['course name']==course, 'room number'] = 'Small-' + str(number)
    number += 1

df.loc[df['status'] ==' not allowed', 'room number']='not assigned'

df


# 쪼렙의 코딩, 대체로 파국이다.
```
