# Q. 실리콘 벨리는 누가 일할까 pt.2
1. 어도비 직원 분포 파이 그래프로 그리기
- 인원이 0인 직군은 그래프에 표시 X

```python
df = pd.read_csv('C:/codeit/silicon_valley_details.csv')
adobe = df['company'] == 'Adobe'
df = df[adobe]
df['job_category'].unique()
```

```python
df = df[['job_category','count']]

condition = (df['count'] != 0)

condition2 = (df['job_category']!='Totals')&(df['job_category']!='Previous_totals')
df = df[condition&condition2]
job = df['job_category'].unique()


# job_list = list(job)
# ex = df[df['job_category']=='Executives']
# ma = df[df['job_category']=='Managers']
# pr = df[df['job_category']=='Professionals']
# sa = df[df['job_category']=='Sales workers']
# ad = df[df['job_category']=='Administrative support']
```

```python
df = pd.read_csv('C:/codeit/silicon_valley_details.csv')
adobe = df['company'] == 'Adobe'
df = df[adobe]
df = df[df['race'] == 'Overall_totals']
zero = df['count']!=0
non_total = (df['job_category']!='Totals')
non_prev = (df['job_category']!='Previous_totals')
df = df[non_total&zero&non_prev]
df.plot(kind = 'pie', y = 'count')
```
