# KDE plot

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('C:/codeit/body.csv', index_col = 0)
```

```python
df['Height'].value_counts().sort_index()
```

```python
df['Height'].value_counts().sort_index().plot()
```

```python
# 위의 그래프는 쫌 힘든데, kde로는 매끄럽게 확인 가능(from seaborn)
sns.kdeplot(df['Height'], bw = 0.5)
# bw는 histogram으로 치면 bins와 같은 것.
```

# 카테고리별 시각화

```python
df = pd.read_csv('C:/codeit/laptops.csv', index_col = 0)
df['os'].unique()
```

```python
sns.catplot(data=df, x= 'os', y= 'price', kind = 'box')
#catplot은 각 os간 비교를 할 수 있도록,(여러 항목끼리 비교할 수 있도록) 도와준다
```

```python
sns.catplot(data=df, x= 'os', y= 'price', kind = 'strip')
# windows는 데이터가 많고, mac은 적음을 알 수 있다.
```

```python
df['processor_brand'].unique()
# 어떤 os가 인텔을 쓰고amd를 쓰는지 알고 싶다
```

```python
sns.catplot(data=df, x= 'os', y= 'price', kind = 'strip', hue = 'processor_brand')
# hue는 색깔을 나타냄
# amd를 쓰는 건 비교적 저렴함을 알 수 있지
```

```python
sns.catplot(data=df, x= 'os', y= 'price', kind = 'swarm', hue = 'processor_brand')
# 점들이 겹쳐보이지 않고 펼쳐진다.
```
