# Q. 서류전형 합격 여부
- 각 파트가 최소 250, 총 점수가 최소 600 이상 되어야 합격
- '합격 여부' column 추가, 합격자는 True, 불합격자는False 불린값 넣어주라

```python
df = pd.read_csv('C:/codeit/toeic.csv')
condition1 = (df['LC']>= 250) & (df['RC']>=250)
condition2 = (df['LC'] + df['RC'] >= 600)
cond = condition1 & condition2

df['합격 여부'] = 'False'

df.loc[cond, '합격 여부'] = 'True'
df

# 요건 내 답변
```

```python
pass_total = df['LC']+df['RC']>600
pass_both = (df['LC'] >= 250) & ( df['RC'] >= 250)
df['합격 여부'] = pass_total & pass_both
df
# 요건 코드잇 답변
# 불린 연산자로 할당해야 했기 때문에, 3번 줄 처럼 코딩을 해줘야 한다.
```
