# 데이터 클리닝: 정확성 pt.2

```python
import pandas as pd
df = pd.read_csv('C:/codeit/beer.csv', index_col = 0 )
df.head()
```

```python
%matplotlib inline
```

```python
df.plot(kind = 'box', y = 'abv')
# 그래프에서 이상점 존재함을 알 수 있다.
```

```python
df['abv'].describe()
```

```python
df['abv'].quantile(0.25)
```

```python
df['abv'].quantile(0.75)
```

```python
q1 = df['abv'].quantile(0.25)
q3 = df['abv'].quantile(0.75)
iqr = q3 - q1
```

```python
condition = (df['abv'] < q1 - 1.5* iqr) | (df['abv'] > q3 + 1.5*iqr)
# 이 조건을 충족하면 이상점이다.
```

```python
df[condition]
# 소주, 보드카는 맥주에 들어가면 안되기에 이상점이 되버린것. 
# 마지막 것은 그냥 잘못 기입되어 이상점이 된 것. 
```

```python
df.loc[2250,'abv'] = 0.055
# 2250row의 abc값을 수정해준 것. 
```

```python
df.loc[2250]
```

```python
condition = (df['abv'] < q1 - 1.5* iqr) | (df['abv'] > q3 + 1.5*iqr)
df[condition]
```

```python
df[condition].index
```

```python
df.drop(df[condition].index, inplace = True)
```

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5*iqr)
df[condition]
```

```python
df.plot(kind = 'box', y = 'abv')
# 이제 깔끔하쥬?
```

# 영화 평점 분석하기 pt.1

```python
df = pd.read_csv('C:/codeit/movie_metadata.csv')
df
```

```python
df.plot(kind = 'box', y = 'budget')
```

```python
df['budget'].describe()
```

```python
q1 = df['budget'].quantile(0.25)
q3 = df['budget'].quantile(0.75)
iqr = q3 - q1
```

```python
con = (df['budget'] > q3 + 10*iqr)

df.drop(df[con].index, inplace = True)
df.plot(kind = 'box', y = 'budget')
```

```python
df.plot(kind= 'scatter', x = 'budget', y = 'imdb_score')
```
