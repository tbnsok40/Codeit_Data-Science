# 서면 1,2호선 대합소, 측정소

```python
import pandas as pd

seomyun2_o = pd.read_csv('C:/codeit/서면2호선측정소.csv')
seomyun2_w = pd.read_csv('C:/codeit/서면2호선대합소.csv')
seomyun1_o = pd.read_csv('C:/codeit/서면1호선측정소.csv')
seomyun1_w = pd.read_csv('C:/codeit/서면1호선대합소.csv')

seomyun2_o.drop([0,1,2,3,4,29],inplace = True)
seomyun2_w.drop([0,1,2,3,4,29],inplace = True)
seomyun1_o.drop([0,1,2,3,4,29],inplace = True)
seomyun1_w.drop([0,1,2,3,4,29],inplace = True)

seomyun2_w.columns = ['시간','PM2.5','PM10','CO2','CO','NO2','NO',"HCHO"]
seomyun2_o.columns = ['시간','PM2.5','PM10','CO2','CO','NO2','NO',"HCHO"]
seomyun1_w.columns = ['시간','PM2.5','PM10','CO2','CO','NO2','NO',"HCHO"]
seomyun1_o.columns = ['시간','PM2.5','PM10','CO2','CO','NO2','NO',"HCHO"]

seomyun2_w.set_index('시간',inplace = True)
seomyun2_o.set_index('시간',inplace = True)
seomyun1_w.set_index('시간',inplace = True)
seomyun1_o.set_index('시간',inplace = True)

#데이터 프레임 역순으로 배열
seomyun1_o[::-1].head()

```
