# 스타트업 아이템 탐색하기

- Branded clothing: 나는 브랜드가 없는 옷보다 브랜드가 있는 옷을 선호한다.
- Healthy eating: 나는 건강하거나 품질이 좋은 음식에는 기쁘게 돈을 더 낼 수 있다.
- Musical instruments: 나는 악기 연주에 관심이 많다.
- New environment: 나는 새 환경에 잘 적응하는 편이다.
- Prioritising workload: 나는 일을 미루지 않고 즉시 해결해버리려고 한다.
- Spending on looks: 나는 내 외모에 돈을 많이 쓴다.
- Workaholism: 나는 여가 시간에 공부나 일을 자주 한다.
- Writing: 나는 시 쓰기에 관심이 많다.
- Writing notes: 나는 항상 메모를 한다.

<가설>
1. 
악기를 다루는 사람들은 시 쓰기를 좋아하는 경향이 있을 것이다.

2. 
외모에 돈을 많이 투자하는 사람들은 브랜드 의류를 선호할 것이다.

3. 
메모를 자주 하는 사람들은 새로운 환경에 쉽게 적응할 것이다.

4. 
워커홀릭들은 건강한 음식을 먹으려는 경향이 있을 것이다.

```python
#1. 타이타닉의 승객은 30대와 40대가 가장 많다. (X)
df.plot(kind = 'hist', y = 'Age', bins = 13)
```

```python
#2. 가장 높은 요금을 낸 사람은 30대이다. (o)
age_fare = df[['Age','Fare']]

sns.catplot(data=age_fare, x= 'Age', y= 'Fare')

#연령대별 Fare의 수치화

df.loc[df['Fare'] > 500,'Age']

# 30대가 가장 많이 낸건 맞는데, 다른 방식으로 추출해보기
```

```python
# 2번을 푸는 다른 방법

df['Fare'].max()
df.loc[df['Fare']==512.3292, 'Age']

# 또 다른 방법

df.plot(kind = 'scatter', x = 'Age', y ='Fare')
```

```python
#3. 생존자가 사망자보다 더 많다. (X)

df['Survived'].value_counts()
# 0 이 사망이므로, 사망자수가 더 많다.
```

```python
#4. 사람이 가장 많은 곳은 3등실 (O)
#sns.catplot(data=df, x = 'Pclass', kind = 'hist')

df['Pclass'].plot(kind = 'hist')
# or
df['Pclass'].value_counts()
```

```python
#5. 생존율 가장 높은 객실은 1등실 (o)

df5 = df.loc[:,'Survived':'Pclass']
df5 = df5[df5['Pclass']==1]
df5['Survived'].value_counts()
```

```python
df5 = df.loc[:,'Survived':'Pclass']
df5 = df5[df5['Pclass']==2]
df5
df5['Survived'].value_counts()
```

```python
df5 = df.loc[:,'Survived':'Pclass']
df5 = df5[df5['Pclass']==3]
df5['Survived'].value_counts()
```

```python
sns.kdeplot(df['Pclass'], df['Survived'])
```

```python
#6. 나이가 어릴수록 생존율이 높다. (O)
df[['Survived','Age']].corr()
```

```python
sns.violinplot(data= df, x = 'Survived', y= 'Age')
# 애초에 어리다는 기준은 무엇... 60대에겐 40대도 어리지
```

```python
#7. 나이보다 성별이 생존율에 더 많은 영향 (X)
df.loc[df['Sex']=='male','Sex']=1
df.loc[df['Sex']=='female','Sex']=0
dff = df[['Survived','Sex','Age']]
sns.clustermap(dff.corr())
# clustermap에서 survived와 age간의 색깔이 더 연하기에, 생존율과 나이가 더 연관성이 깊다.
```
